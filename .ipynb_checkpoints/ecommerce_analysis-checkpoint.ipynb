```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv('ecommerce_data.csv')
```

```python
 #1. Sales Trend Analysis Over Time
df['Order Date'] = pd.to_datetime(df['Order Date'])
sales_trend = df.groupby(df['Order Date'].dt.to_period('M')).sum(numeric_only=True)['Price']
```

```python
plt.figure(figsize=(12, 6))
sales_trend.plot(kind='line', marker='o', color='b')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.grid(True)
plt.show()
```

```python
# 2. Product Popularity (Top 10 Products by Quantity Sold)
product_popularity = df.groupby('Product')['Quantity'].sum().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 6))
product_popularity.plot(kind='bar', color='orange')
plt.title('Top 10 Most Popular Products')
plt.xlabel('Product')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# 3. Sales by Location
sales_by_location = df.groupby('Location').sum(numeric_only=True)['Price']

```

```python
plt.figure(figsize=(12, 6))
sales_by_location.plot(kind='bar', color='green')
plt.title('Sales by Location')
plt.xlabel('Location')
plt.ylabel('Total Sales ($)')
plt.grid(True)
plt.show()
```

```python
# 4. Sales Performance by Category
category_performance = df.groupby('Category').sum(numeric_only=True)['Price']
```

```python
plt.figure(figsize=(12, 6))
category_performance.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['skyblue', 'orange', 'lightgreen', 'pink', 'yellow', 'purple'])
plt.title('Sales Performance by Category')
plt.ylabel('')
plt.show()
```
